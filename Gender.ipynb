```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, plot_confusion_matrix
from sklearn.model_selection import GridSearchCV
from  xgboost import XGBClassifier
from sklearn import metrics
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
col_names = []
for col in data.columns:
    col_names.append(col)
```

```python
data_1 = data[col_names[1:12]]
```

```python
data_2 = data[col_names[12:]].rename(columns={col_names[12]: col_names[1], 
                                   col_names[13]: col_names[2],
                                  col_names[14]: col_names[3],
                                  col_names[15]: col_names[4],
                                  col_names[16]: col_names[5],
                                  col_names[17]: col_names[6],
                                  col_names[18]: col_names[7],
                                  col_names[19]: col_names[8],
                                  col_names[20]: col_names[9],
                                  col_names[21]: col_names[10],
                                  col_names[22]: col_names[11]}, errors="raise")
```

```python
frames=[data_1, data_2]

final = pd.concat(frames)
final
```

```python
final[['nose_wide', 'nose_long', 'lips_thin', 'distance_nose_to_lip_long', 'long_hair']] = final[['nose_wide', 'nose_long', 'lips_thin', 'distance_nose_to_lip_long', 'long_hair']].round()
final
```

```python
final['gender'].value_counts()
```

```python
final.isnull().sum()
```

```python
final = final[final['long_hair'].notnull()]
```

```python
final.isnull().sum()
```

```python
pred_x = final[:1001].drop(columns = 'gender')
pred_x
```

```python
x = final[1001:].drop(columns = 'gender')
y = final[1001:]['gender'].replace({"Female":1,"Male":0})
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)
```

```python
model = KNeighborsClassifier()
k_range = list(range(1, 100))
param_grid = dict(n_neighbors=k_range)
  
grid = GridSearchCV(model, param_grid, cv=10, scoring='accuracy', return_train_score=False,verbose=1)
  
grid_search=grid.fit(x_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
accuracy = grid_search.best_score_ *100
print("Accuracy for our training dataset with tuning is : {:.2f}%".format(accuracy) )
```

```python
model = XGBClassifier(learning_rate=0.01).fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
pr, rc, fs, sup = metrics.precision_recall_fscore_support(y_test, y_pred, average='macro')
res = {"Accuracy": round(metrics.accuracy_score(y_test, y_pred), 4),
                              "Precision": round(pr, 4), "Recall":round(rc, 4), "FScore":round(fs, 4)}
res
```
